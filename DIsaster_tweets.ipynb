```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'/train.csv')
df
```

```python
# Lower casing
df['text']=df['text'].str.lower()
df
```

```python
# Remove HTML tags
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
df['text']=df['text'].apply(remove_html_tags)
df
```

```python
# Removing URL.
def remove_url(text):
    pattern = re.compile(r'http?://\S+|www\.\S+')
    return pattern.sub(r'',text)
df['text']=df['text'].apply(remove_url)
df
```

```python
# Converting EMoji into unicode

import string
string.punctuation
exclude = string.punctuation
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
df['text']= df['text'].apply(remove_punc)
df
```

```python
file = '/slang.txt'

# open the file in read mode
with open(file, 'r') as file:
    # read lines from the file
    lines = file.readlines()

# initialize an empty dictionary
res = {}

# iterate through each line and split key-value pairs
for line in lines:
    # strip whitespace and skip empty lines
    line = line.strip()
    if not line or ':' not in line:
        continue  # skip invalid lines

    # split into key and value safely
    parts = line.split(':', 1)  # split at the first colon only
    key = parts[0].strip()
    value = parts[1].strip()
    res[key] = value

# print the dictionary
print("Dictionary Created")
print(res)
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in abbreviations:
            new_text.append(abbreviations[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

df['text'] = df['text'].apply(chat_conversion)
```
